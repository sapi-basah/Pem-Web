```python
import matplotlib.pyplot as plt
import pandas as pd

# Meload Dataset
df = pd.read_csv('/content/drive/MyDrive/Big Data/File csv/heart_statlog_cleveland_hungary_final (1).csv')

#menghitung categori kolom umur
Age_counts = df['age'].value_counts()

# Membuat bar plot
plt.figure(figsize=(10, 5))
plt.bar(Age_counts.index, Age_counts.values )

# menambahkan label dan judul pada sumbu x dan sumbu y
plt.xlabel('Umur')
plt.ylabel('Jumlah')
plt.title('Distribusi dalam kolom umur')

plt.xticks(rotation=45, ha='right') # Merotate title pada sumbu x

plt.tight_layout
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Meload Dataset
# Jika anda belum memuatnya di sel sebelumnya, hapus komentar pada baris di bawah:
# df = pd.read_csv('/content/drive/MyDrive/Big Data/File csv/heart_statlog_cleveland_hungary_final (1).csv')

# Membuat figure untuk plot
plt.figure(figsize=(10, 5)) # Sesuaikan ukuran figure
plt.scatter(df['cholesterol'], df['max heart rate'])

plt.xlabel('Nilai Cholesterol')
plt.ylabel('Nilai Detak Jantung')
plt.title('Scatter Plot Cholesterol vs Detak jantung')

# Mengatur rotasi label sumbu x jika diperlukan
plt.xticks(rotation=45, ha='right') # Merotate title pada sumbu x

plt.tight_layout
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# mencari nilai yang bernilai angka untuk melihat korelasi antar kolom
numeric_df = df.select_dtypes(include=['number'])
numeric_df.head()

correlation_matrix = numeric_df.corr()

sns.set_style("whitegrid")

plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix,
    annot=True,      # Menampilkan nilai korelasi pada peta panas
    cmap='coolwarm', # Pilih peta warna (dingin/hangat baik untuk menunjukkan korelasi positif/negatif)
    linewidths=0.5,  # Memformat anotasi menjadi 2 angka di belakang koma
    fmt=".2f"        # Menambahkan garis di antara sel
)

plt.title("Korelasi Antar Kolom")

plt.tight_layout()
plt.show()
```
